````python
def trainer_function(train_dataloader, model, error_func, optimizer):
    for batch_no, example in enumerate(train_dataloader):
        image_tensors  = example['image']
        y_actual       = example['label']

        y_predicted_logits = model.forward(image_tensors)
        loss = error_func(y_predicted_logits.to(torch.float32), y_actual.to(torch.long))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        
    
````
