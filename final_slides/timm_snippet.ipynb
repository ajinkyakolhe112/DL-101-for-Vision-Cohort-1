```python
import torch
import timm # Torch Image Models 

batch_random = torch.randn(2, 3, 224, 224)

model                   = timm.create_model('resnet50', pretrained=True)

class Transfer_Learning(torch.nn.Module):
    def __init__(self):
        self.model       = timm.create_model('resnet50', pretrained=True) # FREEZING LAYERS. 
        self.new_layer   = torch.nn.Conv2d(in_channels = 512, out_channels = 10, kernel_size=(3,3))
    def forward(self, x_batch):
        output = self.model(x_batch)
        output = self.new_layer(output)

```
